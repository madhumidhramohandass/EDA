```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/content/Superrstore.csv")
```

```python
df.head(5)
```

```python
df.drop(columns="Postal Code",inplace=True)
```

```python
df.head(3)
```

```python
print(df["Ship Mode"].unique())
print(df["Segment"].unique())
print(df["Country"].unique())
print(df["Category"].unique())
print(df["Sub-Category"].unique())
print(df["Region"].unique())
```

```python
df.info()
```

```python
#checking if there is any null value
df.isna().sum()
```

```python
#sales analysis based on region
df.groupby("Region")["Sales"].sum().plot.bar(width=0.7,color="Teal")
```

```python
#profit analysis based on region
df.groupby("Region")["Profit"].sum().plot.bar(width=0.5,color="maroon")
```

```python
#sales analysis based on region - piechart
# to see values within pie autopct=%1.0f%%

#df.groupby("Region")["Sales"].sum().plot.pie()

df.groupby("Region")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Profit analysis based on region
df.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#sales analysis based on segment
df.groupby("Segment")["Sales"].sum().plot.bar(color="violet")
```

```python
#Profit analysis based on segment
df.groupby("Segment")["Profit"].sum().plot.bar(color="grey")
```

```python
#sales analysis based on category
df.groupby("Category")["Sales"].sum().plot.bar(color="purple")
```

```python
#Profit analysis based on category
df.groupby("Category")["Profit"].sum().plot.bar(color="indigo")
```

```python
#sales analysis based on category
df.groupby("Category")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Profit analysis based on category
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#sales analysis based on state
df.groupby("State")["Sales"].sum().plot.bar(width=0.3)
```
